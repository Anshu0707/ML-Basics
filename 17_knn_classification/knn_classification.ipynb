# <span style="color:purple; font-weight:bold">K Nearest Neighbor 

Supervised Machine Learning Algorithm used for Classification and Regression.               
Mechanism: Classifies new points based on the proximity (similarity) of surrounding data points.                    
Model Building: "Lazy learner"—no explicit training phase; memorizes all data points.                   

###  <span style="color:purple; font-weight:bold">Key Concepts</span>                     



- <span style="color:blue; font-weight:bold">K Value:</span> Number of neighbors to consider (user-defined integer).      

    * <span style="font-weight:bold">Tip:</span> Use an odd K to avoid ties in classification votes.     

- <span style="color:blue; font-weight:bold">Distance Metrics:</span> Mathematical formulas to measure similarity (e.g., Euclidean, Manhattan distance).                

- <span style="color:blue; font-weight:bold">Decision Rule:</span> Majority vote among the K nearest neighbors determines the new point's class.          

- <span style="color:blue; font-weight:bold">Scaling:</span> Requires feature scaling (normalization/standardization) because distance is sensitive to data magnitude.     

- <span style="color:blue; font-weight:bold">Pros/Cons:</span> Simple, easy to implement. Slow for large datasets and sensitive to noisy data/outliers.

## <span style="color:purple; font-weight:bold">Distance Metrics in KNN                     

For calculating distances KNN uses various different types of distance metrics. For the algorithm to work efficiently, we need to select the most appropriate distance metric. Some of these metrics are as mentioned below:                        


1. <span style="color:blue; font-weight:bold">Euclidean Distance:</span> This is the default metric used by the sklearn KNN classifier. It is often referred to as L2 Norm. It calculates the ordinary straight line distance between two points in a Euclidean space.

2. <span style="color:blue; font-weight:bold">Manhattan Distance:</span> Often referred to as the city block metric, in this the distance between two points is the absolute differences of their Cartesian coordinates and is used in case of high dimensionality.                       

3. <span style="color:blue; font-weight:bold">Hamming Distance:</span> This metric is used for comparing two binary data strings and is typically used with OneHotEncoding. It looks at the entire dataset and discovers when data points are similar or dissimilar.

# <span style="color:blue; font-weight:bold">A few thoughts on picking a value for “K”</span>

- There is no physical or biological way to determine the best value for “K”, so you may have to try out a few values before settling on one.  
  Do this by pretending part of the training data is “unknown”.

- Low values for K (like **K = 1** or **K = 2**) can be noisy and subject to the effects of outliers.

- Large values for K smooth over things, but you don’t want K to be too large; otherwise a category with only a few samples in it will always be out-voted by other categories.

```python
import pandas as pd
from sklearn.datasets import load_iris
iris = load_iris()
```

```python
iris.feature_names
```

```python
iris.target_names
```

```python
df = pd.DataFrame(iris.data, columns=iris.feature_names)
df.head()
```

```python
df['target'] = iris.target
df.head()
```

```python
df[df.target==1].head()
```

```python
df[df.target==2].head()
```

```python
df['flower_name'] = df.target.apply(lambda x:iris.target_names[x])
df.head()
```

```python
df[45:55]
```

```python
df0 = df[:50]
df1 = df[50:100]
df2 = df[100:]
```

```python
import matplotlib.pyplot as plt
%matplotlib inline
```

**Sepal Length vs Sepal Width [Setosa vs Versicolor]**

```python
plt.title('Petal Length vs Petal Width [Setosa vs Versicolor]')
plt.xlabel('Sepal Length')
plt.ylabel('Sepal Width')
plt.scatter(df0['sepal length (cm)'], df0['sepal width (cm)'], color = 'green', marker = '+')
plt.scatter(df1['sepal length (cm)'], df1['sepal width (cm)'], color = 'blue', marker = '.')
```

**Petal Length vs Petal Width [Setosa vs Versicolor]**

```python
plt.title('Petal Length vs Petal Width [Setosa vs Versicolor]')
plt.xlabel('Petal Length')
plt.ylabel('Petal Width')
plt.scatter(df0['petal length (cm)'], df0['petal width (cm)'], color = 'green', marker = '+')
plt.scatter(df1['petal length (cm)'], df1['petal width (cm)'], color = 'blue', marker = '.')
```

```python
# Train-Test Split
from sklearn.model_selection import train_test_split

X = df.drop(['target', 'flower_name'], axis = 'columns')
y = df.target 
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=1)
```

```python
print(f"Length of X_train: {len(X_train)} \nLength of X_test: {len(X_test)}")
```

**Let's use KNN model now**

```python
from sklearn.neighbors import KNeighborsClassifier
knn = KNeighborsClassifier(n_neighbors=10)
```

```python
knn.fit(X_train, y_train)
```

```python
knn.score(X_test, y_test)
```

```python
knn.predict([[4.8, 3.0, 1.5, 0.3]])
```

**Let's plot confusion matrix for the model**

```python
from sklearn.metrics import confusion_matrix
y_pred = knn.predict(X_test)

cm = confusion_matrix(y_test, y_pred)
cm
```

```python
%matplotlib inline
import matplotlib.pyplot as plt
import seaborn as sn

plt.figure(figsize=(7, 5))
sn.heatmap(cm, annot=True)
plt.xlabel('Predicted')
plt.ylabel('Truth')
```

**Let's print classification report for precision, recall and f1-score for each classes**

```python
from sklearn.metrics import classification_report

print(classification_report(y_test, y_pred))
```
