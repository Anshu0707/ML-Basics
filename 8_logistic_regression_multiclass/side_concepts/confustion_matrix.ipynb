<h1 style='color:purple; font-weight:bold'> CONFUSION MATRIX </h1>

1. The confusion matrix is a 2 dimensional array comparing predicted category labels to the true label.
2. For binary classification, these are the <e style="color:purple">True Positive, True Negative, False Positive and False Negative categories</e>.    


![alt text](../assets/images/Model_Evaluation.png)

---

Python provides its confusion_matrix(y, y_pred) function from sklearn.
The output is a 2-D Array, to make the output beautiful we use <e style="color:purple">heatmap from the Seaborn library</e>.

A confusion matrix will always be a square matrix, and the ranges (or list of class labels) for both the X-axis and Y-axis are always the sam
Each cell of the array(graph) has (x,y) coordinate. 

X-axis --> label of values for what the model predicted    
Y-axis --> label of values for what the actual value was

![alt text](../assets/images/Confusion_Matrix.png)    


1. At x='Dog', y='Cat' the cell(x,y) = 2

    This means that were 2 times when the actual value was 'Cat' but model predicted 'Dog'. This highlights inaccuracy.

2. At x='Dog', y='Dog' the cell(x,y) = 10

    This means that were 10 times when the actual value was 'Dog' and the model predicted 'Dog'. This signifies accuracy.    


- So for model to be accurate predicted(y) value needs to be equal to actual(x) value.    
- In the graph only the <e style="color:purple">principal diagonal cells</e> are the one where x=y condition meets.    
- So all the <e style="color:purple">non diagonal cells where (x,y) > 0</e> its signifying model inaccuray(specifically the frequncy of wrong predictions).

---


<h2 style="color:purple; font-weight:bold">Conclusion ?</h2>

1. Diagonal cells indicate True Positives and True Negatives    
   Non-diagonal cells indicate False Positives and False Negatives

2. <e style="color:purple">Diagonal cells have large values</e> --> Model Accuracy High     
   <e style="color:purple">Non Diagnoal cells have large values</e> --> Model Accuracy Low

```python
import numpy as np
import pandas as pd
from sklearn.metrics import confusion_matrix
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline

# Example data 
y_true = ["Cat", "Dog", "Cat", "Dog", "Cat", "Cat"]   # Actual Data
y_pred = ["Cat", "Cat", "Cat", "Dog", "Dog", "Cat"]   # Predicted Data

# 1. Computing the confusion matrix
cm = confusion_matrix(y_true, y_pred, labels=["Cat", "Dog"])

# 2. Create a Pandas DataFrame for better labeling
df_cm = pd.DataFrame(cm, index=["Actual Cat", "Actual Dog"], columns=["Predicted Cat", "Predicted Dog"])

# 3. Plot the heatmap
plt.figure(figsize=(7, 6))
sns.heatmap(df_cm, annot=True, fmt='d', cmap='Blues') # annot=True shows the numbers, fmt='d' ensures integer formatting
plt.xlabel("Predicted labels")
plt.ylabel("Actual labels")
plt.title("Confusion Matrix")
plt.show()
```
