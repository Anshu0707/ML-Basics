```python
import pandas as pd
import numpy as np
from sklearn import linear_model
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('./assets/files/homeprices.csv')
df
```

```python
%matplotlib inline
plt.xlabel('area')
plt.ylabel('price')
plt.scatter(df.area,df.price,color='red',marker='+')
```

```python
new_df = df.drop('price',axis='columns')
new_df
```

```python
price = df.price
price
```

```python
# Create linear regression object
reg = linear_model.LinearRegression()
reg.fit(new_df,price)
```

```python
reg.predict([[3300]])
```

```python
reg.coef_
```

```python
reg.intercept_
```

```python
reg.predict([[5000]])  #  Predict price of a home with area = 5000 sqr ft
```

<span style="color:purple; font-weight:bold">Generate CSV file with list of home price predictions

```python
area_df = pd.read_csv("./assets/files/areas.csv")
area_df.head(3)
```

```python
p = reg.predict(area_df)
p
```

```python
area_df['prices']=p
area_df
```

```python
area_df.to_csv("prediction.csv")
```
