```python
import pandas as pd
import numpy as np
from sklearn import linear_model
```

```python
df = pd.read_csv("./assets/csv_files/homeprices.csv")
df
```

** Data PreProcessing **

```python
df.bedrooms.median()
```

```python
df.bedrooms = df.bedrooms.fillna(df.bedrooms.median())
df
```

```python
reg = linear_model.LinearRegression()
reg.fit(df.drop('price',axis='columns'),df.price)
```

```python
reg.coef_
```

```python
reg.intercept_
```

```python
reg.predict([[3000, 3, 40]]) ;
```
