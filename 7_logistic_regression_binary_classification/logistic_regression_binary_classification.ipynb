# <span style="color:purple; font-weight:bold"> Logistic Regression [Binary Classification]

<span style="font-weight:bold">For the predictions where the <ins style="color:blue">output</ins> is a <ins style="color:blue">categorical outcome</ins> like '0'/'1', 'True'/'False', 'Yes'/'No' we use Logistic Regression (Binary Classification) as Linear Regression fails for these type of outcomes.</span>

### Logistic Regression uses <ins style="color:blue"> Sigmoid function (S-shaped curve)</ins> to better fit the line.

---



---

<span style="color:blue; font-weight:semibold"> You can see for categorical outcomes Linear Regression doesn't fit the best line.



---

```python
import pandas as pd 
import matplotlib.pyplot as plt 
%matplotlib inline
```

```python
df = pd.read_csv('./assets/files/insurance_data.csv')
df.head()
```

```python
plt.scatter(df.age, df.bought_insurance, marker='+', color='red')
```

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(df[['age']], df.bought_insurance, test_size=0.1)    
```

```python
X_test
```

```python
from sklearn.linear_model import LogisticRegression
model = LogisticRegression()
model.fit(X_train, y_train)
```

```python
X_test
```

```python
y_predicted = model.predict(X_test)
```

```python
model.predict_proba(X_test)  # Left column is probability of class 0, right column is probability of class 1 for each entry in X_test
```

```python
y_predicted # 0 means the person won't buy insurance, 1 means will buy insurance
```

```python
X_test
```

```python
model.coef_         # This indicates the value of 'm' in y= m*x + b equation.
```

```python
model.intercept_    # This indicates the value of 'b' in y= m*x + b equation.
```

# <span style="color:purple;">Here I am implementing Sigmoid function by doing math and without using Logistic Regression module of python.

<span style="color:purple;">**I will be using same <ins style="color:blue;">model.coef_</ins> and <ins style="color:blue;">model.intercept_</ins> values predicted by LogisticRegression model as 'm' and 'b' value in my sigmoid function.**

```python
import math
def sigmoid(x):
  return 1 / (1 + math.exp(-x))

def predict(age):
  z = 0.042 * age - 1.53
  y = sigmoid(z)
  return y
```

```python
age = 43
predict(age)    # Value greater than 0.5 indicates the person is likely to buy insurance which is the same prediction we got from the model.predict method earlier.
```
